# M2M100Transformer

The M2M100Transformer introduces a novel approach to multilingual translation. By supporting direct translation across 9,900 language pairs from 100 languages, it marks a significant advancement in overcoming language barriers for global communication.


## Overview of M2M100

The M2M100 model is a leap forward in multilingual machine translation, designed to translate directly between any pair of languages from a set of 100. Unlike previous translation models that were predominantly English-centric, the M2M100 model focuses on a truly Many-to-Many translation approach, free from English-centric biases. This innovation is made possible by a meticulously crafted dataset covering thousands of language directions with supervised data, sourced through extensive large-scale mining.

## Key Features

- **True Multilingual Translation**: Direct translation across 9,900 language pairs without relying on English as an intermediary language.
- **Innovative Training Dataset**: A new training approach with a dataset covering supervised data for thousands of language directions, moving beyond English-centric data.
- **Enhanced Model Capacity**: Through dense scaling and language-specific sparse parameters, the model achieves high-quality translation outputs.
- **Non-English-Centric Approach**: A focused effort on improving translation quality between non-English language pairs, demonstrating gains of over 10 BLEU points in direct translations.

## M2M100Transformer

Leveraging the power of M2M100, the M2M100Transformer annotator in Spark NLP is specially designed for efficient, high-quality multilingual translation tasks. It embodies the spirit of global communication, enabling direct and fluent translations between a vast array of languages.
